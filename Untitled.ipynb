```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 

from sklearn import metrics 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, DecisionTreeRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

#### Cargamos el DataSet y Mostramos las primeras 5 Filas ####

```python
Datos = pd.read_csv("Titanic-Dataset.csv")
Datos.head()
```

#### Cantidad de Datos y Variables ####

```python
Datos.shape
```

#### Cantidad de Datos, teniendo en cuenta la media, percetiles datos minimos y maximos ####

```python
Datos.describe()
```

#### Busqueda de datos nulos ####

```python
Datos.isna().sum()
```

#### Limpiamos los Datos ####

```python
# Rellenar los valors faltantes de la columna "Age" con el promedio
Datos['Age'] = Datos['Age'].fillna(Datos['Age'].mean())

# Eliminamos la columna "Cabin", debido a que falta una gran cantidad de valores
Datos = Datos.drop(columns=['Cabin'])

# Rellenar los valores faltantes de la columna "Embarked" con la moda
Datos['Embarked'] = Datos['Embarked'].fillna(Datos['Embarked'].mode()[0])

# Verificar si aún hay valores faltantes
Datos.isna().sum()
```

#### Clasificación de pasageros en la variable Age ####

```python
plt.hist(Datos.Age)
```

#### Analizamos la corelacion entre variable y la calidad, teniendo en cuenta que 0 es que no sobrevivio y 1 que si sobrevivio ####

```python
sns.scatterplot(x="Age", y="Sex", hue="Survived", data=Datos)
```

#### Analizamos los datos de las variables que se usaron para entrenar y testear el modelo ####

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=Datos[['Survived', 'Age', 'Pclass', 'Parch']])
plt.xticks(rotation=45)
plt.title('Distribición de Variables del Titanic')
plt.ylabel('Valores')
plt.xlabel('Variables')
plt.show()
```

#### Buscamos los datos antipicos en la variable "Age" ####

```python
Edad = (Datos['Age']>300).sum()
print('Datos antipicos de la variable Age: ', Edad)
```

#### Establecer las variables almacenadas, que contengan los datos que entrenan y testean el modelo #### 

```python
base_predicciones=['Survived', 'Age', 'Pclass', 'Parch']
variable_a_predecir=['Age']
```

```python
predicciones = Datos[base_predicciones]
predecir = Datos[variable_a_predecir]
```

```python
predicciones
```

```python
predecir
```

#### Miramos la corelación entre variables ####

```python
correlacion = predicciones.corr()
correlacion
```

```python
sns.heatmap(correlacion)
plt.title('Matriz de Correlación')
plt.show()
```

#### Creamos el modelo, y seleccionar la cantidad de datos usados para entrenar y testear el modelo ####

```python
x_train, x_test, y_train, y_test = train_test_split(predicciones, predecir, test_size = 0.15, random_state = 13)
```

```python
arbol = DecisionTreeRegressor()
```

```python
arbol.fit(x_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones=arbol.predict(x_test)
```

```python
predicciones
```

#### Examinamos la cantidad de Datos en una matriz de confución ####

#### Esta matriz se encarga de comparar los datos reales con los predichos por el medelo ####

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames = ['Actual'], colnames = ['Predicciones'])
```

#### Ahora evaluamos el rendimiento del modelo #### 

```python
mae = mean_absolute_error(y_test, predicciones)
print(f"Error Absoluto Medio (MAE): {mae}")
```

#### Al evaluar con el modelo "mae": ####
#### indica que, en promedio, las diferencias entre las predicciones del modelo y los valores reales son prácticamente nulas. Esto refleja un nivel de precisión extremadamente alto en las estimaciones realizadas por el modelo. El modelo es altamente efectivo en minimizar los errores absolutos entre las predicciones y los datos reales, lo que asegura que las predicciones estén muy cerca de los valores observados ####

```python
mse = mean_squared_error(y_test, predicciones)
print(f"Error Cuadrático Medio (MSE): {mse}")
```

#### Tras evaluar con el modelo "mse": ####
#### Evidenciamos que los errores al cuadrado, que penalizan desviaciones más grandes, también son insignificantes. Este valor extremadamente bajo refuerza que las diferencias entre las predicciones y los valores reales son casi inexistentes. El modelo no solo es preciso, sino que también es robusto, ya que incluso los errores mayores (si existieran) son prácticamente eliminados ####

```python
r2 = r2_score(y_test, predicciones)
print(f"R²: {r2}")
```

#### Al evaluar con el modelo "r2":
#### El R² igual a 1.0 significa que el modelo explica el 100% de la variabilidad en los datos de sobrevivencia del Titanic. Esto indica que las predicciones se ajustan perfectamente a los valores reales. El modelo logra un ajuste perfecto, lo que lo convierte en una herramienta ideal para predecir la cantidad de sobrevivientes basada en las características utilizadas durante el entrenamiento ####
